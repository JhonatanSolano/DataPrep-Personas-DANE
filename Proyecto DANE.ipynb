## **2. Definición del diseño y/o arquitectura de la solución**
---



### **2.1. Descripción**
---


La solución propuesta se estructura como un sistema de **análisis Big Data** orientado a transformar los microdatos del Sisbén IV en información geoespacial útil para la toma de decisiones, combinando un *pipeline* **ETL distribuido**, un **modelo documental flexible** y un motor de **visualización interactivo**. La arquitectura sigue un enfoque modular que facilita la mantenibilidad, la escalabilidad y la separación de responsabilidades técnicas dentro del sistema.

---

### Estructura General de la Solución

La arquitectura se organiza en **tres capas principales**:

| Capa | Herramienta Clave | Propósito |
| :--- | :--- | :--- |
| **Ingesta y Preparación** | Apache Spark (PySpark) | Implementa el ETL distribuido para procesar $4.46$ millones de registros en paralelo. |
| **Persistencia** | MongoDB (NoSQL) | Almacena los indicadores agregados en un modelo flexible (Departamento $\to$ Municipio $\to$ Zona). |
| **Análisis y Visualización** | BI Geoespacial (Power BI / Tableau) | Consume los datos para generar mapas de calor, tablas de priorización y análisis de correlación. |

#### Capa de Ingesta y Preparación (ETL Distribuido)

* Se implementa en **Apache Spark** utilizando **DataFrames** para leer, validar y transformar los $4.46$ millones de registros.
* Incluye **limpieza** de variables de privación ($I1-I15$), **normalización** de códigos territoriales (Divipola), **conversión** de tipos y **cálculo del Índice de Pobreza Multidimensional (IPM)** ponderado por el Factor de Expansión (**FEX**).
* Esta etapa permite procesar en memoria grandes volúmenes de datos de forma paralela, reduciendo tiempos de ejecución de horas a minutos.

#### Capa de Persistencia (Modelo NoSQL con MongoDB)

* Los indicadores se almacenan en colecciones organizadas bajo la jerarquía: **Departamento $\to$ Municipio $\to$ Zona $\to$ Indicadores de pobreza.**
* Se utiliza **MongoDB** por su capacidad para manejar documentos flexibles y **consultas geoespaciales eficientes**, permitiendo agregar nuevas variables en futuras versiones del Sisbén sin alterar un esquema rígido.

#### Capa de Análisis y Visualización (Dashboard geoespacial)

* Los datos depurados son consumidos desde **MongoDB** por una capa visual que genera **mapas de calor**, **tablas de priorización** y **análisis de correlación** (ej. "piso inadecuado" vs. enfermedades respiratorias).
* Este panel permite a entidades del Estado **focalizar intervenciones con precisión submunicipal** (cabecera, rural disperso, zonas críticas).

---

### Componentes Clave de la Solución

1.  **Apache Spark (PySpark):**
    * Motor de procesamiento distribuido para manipulación del *dataset* masivo.
    * Permite ejecutar cálculos como IPM, agregaciones, cruces entre variables y análisis estadísticos en pocos segundos.
    
2.  **Pipeline ETL Automatizado:**
    * Orquesta la carga, limpieza, validación y transformación.
    * Minimiza errores humanos y hace el sistema **reutilizable** para futuras actualizaciones de datos.
3.  **MongoDB:**
    * Base **NoSQL** orientada a documentos, ideal para almacenar resultados agregados con estructura variable.
    * Facilita consultas rápidas por ubicación geográfica (`cod_mpio`, `zona`) y exploración interactiva en *dashboards*.
4.  **Herramientas de Visualización (Power BI / Tableau / Folium / Kepler.gl):**
    * Conectan directamente a la base MongoDB, permitiendo la producción de **mapas coropléticos** y análisis territoriales.

#### **¿Cuáles son las tecnologías seleccionadas para el proyecto?**
Se utilizarán Apache Spark (librería PySpark) para el procesamiento de datos y MongoDB para la persistencia (almacenamiento) de la información procesada.

---

#### **¿Por qué se eligieron estas tecnologías en lugar de otras alternativas?**

- Spark vs. Pandas: Se eligió Spark porque herramientas como Pandas (en una sola máquina) presentan problemas de memoria y rendimiento al intentar cargar y manipular un dataset de 4.46 millones de registros.

- MongoDB vs. SQL Relacional: Se eligió MongoDB por su flexibilidad para almacenar datos jerárquicos y su capacidad de escritura rápida, evitando la rigidez de los esquemas relacionales tradicionales durante la fase de carga masiva.

---

#### **¿Cómo se alinean las tecnologías seleccionadas con los objetivos del proyecto?**

El objetivo principal es gestionar un gran volumen de datos de manera eficiente. Spark cumple esto permitiendo la ingesta y limpieza distribuida de los millones de registros del Sisbén, mientras que MongoDB asegura que los datos limpios estén disponibles en una estructura escalable para consultas futuras.

---

#### **¿Cuáles son las ventajas clave de cada tecnología seleccionada?**

- Spark: Procesamiento en memoria (RAM), lo que acelera drásticamente las tareas de limpieza y transformación (ETL) frente a la lectura en disco.

- MongoDB: Esquema dinámico que permite iterar rápidamente en el diseño de la base de datos sin necesidad de migraciones complejas.

---

#### **¿Existen desventajas o limitaciones que deben ser abordadas?**

- Spark: Alto consumo de memoria RAM. Mitigación: Se configurará la sesión de Spark limitando el uso de recursos en el entorno de Google Colab.

- MongoDB: No es eficiente para realizar cruces (joins) complejos entre colecciones. Mitigación: Todos los cruces y unificaciones de datos se realizarán previamente en Spark antes de guardar el resultado final en Mongo.

### ¿Cómo se justifica la necesidad de cada paso?

El preprocesamiento se limitó a dos acciones estratégicas, justificadas por las siguientes necesidades:

    * En la fase de preprocesamiento con el **Feature Engineering**, cuyo objetivo principal fue optimizar el conjunto de datos para garantizar la eficiencia operativa.
2.  **Recodificación Binaria de `PER008` (Salud):**
    * **Justificación:** La variable original estaba en la escala `{1, 2}`. Se transformó a la escala **binaria `{1, 0}`** (donde 1 = Sí / Enfermedad, 0 = No / Sin Enfermedad).

```python
print("--- Ejecutando Transformación Semántica ---")

# Mapeo de Edad (Ciclo Vital)
df_etiquetado = df_analisis.withColumn(
    "Ciclo_Vital",
    when(col("PER002") == 1, "01. Primera Infancia (0-5)")
    .when(col("PER002") == 2, "02. Infancia (6-11)")
    .when(col("PER002") == 3, "03. Adolescencia (12-17)")
    .when(col("PER002") == 4, "04. Juventud (18-28)")
    .when(col("PER002") == 5, "05. Adultez (29-59)")
    .when(col("PER002") == 6, "06. Adulto Mayor (>60)")
    .otherwise("Desconocido")
)

# Transformación de Salud (PER009 - Acceso) a Binario
# Original: 1=Si, 2=No, 9=No aplica
# Nuevo: 1=Accedió, 0=No Accedió (incluye los que no aplicaban)
df_salud_binario = df_etiquetado.withColumn(
    "Acceso_Salud_Binario",
    when(col("PER009") == 1, 1).otherwise(0)
)
df_final = df_salud_binario.withColumn(
    "Enfermedad_Binaria",
    when(col("PER008") == 1, 1).otherwise(0)
)

print("Visualización tabla final:")
df_final.show(5, truncate=False)
```

## **2. Estrategia de Carga Distribuida**

Para garantizar la eficiencia en la subida a la nube, se optó por una estrategia de **inserción paralela a nivel de partición** (`foreachPartition`). En lugar de centralizar los datos en un solo nodo (lo que causaría cuellos de botella), cada trabajador de Spark establece su propia conexión a MongoDB y transmite su porción de datos de manera independiente.

```python
import pymongo
from pymongo import MongoClient
import certifi

realizar_carga = True

# VERIFICACIÓN PREVIA
try:
    client_main = MongoClient(uri, tlsCAFile=certifi.where())
    db_main = client_main["SisbenDB"]
    col_name = "Personas_Procesadas"

    if col_name in db_main.list_collection_names():
        conteo_actual = db_main[col_name].count_documents({})
        if conteo_actual > 0:
            print(f"La colección '{col_name}' ya existe y contiene {conteo_actual:,} documentos.")
            print("Proceso abortado: No se borrarán ni añadirán nuevos datos.")
            realizar_carga = False
        else:
            print("La colección existe pero está vacía. Se procederá con la carga.")
    else:
        print("La colección no existe. Se creará y cargará.")

    client_main.close()

except Exception as e:
    print(f"Error verificando MongoDB: {e}")
    realizar_carga = False

def subir_particion_optimizado(iterator):
    try:
        client_worker = MongoClient(uri, tlsCAFile=certifi.where())
        db = client_worker["SisbenDB"]
        collection = db["Personas_Procesadas"]

        batch = []
        BATCH_SIZE = 2000

        for row in iterator:
            batch.append(row.asDict())

            if len(batch) >= BATCH_SIZE:
                collection.insert_many(batch, ordered=False)
                batch = []

        if batch:
            collection.insert_many(batch, ordered=False)

    except Exception as e:
        print(f"Error en worker: {e}")

    finally:
        client_worker.close()

if realizar_carga:
    try:
        total_docs = df_mongo.count()
        print(f"Iniciando carga de {total_docs:,} documentos...")
        df_mongo.rdd.foreachPartition(subir_particion_optimizado)
        print("Carga finalizada correctamente.")
    except Exception as e:
        print(f"Error durante la carga: {e}")
else:
    print("Verificación completada. Los datos en MongoDB se mantuvieron intactos.")
```

## **4. Ejemplo de Consulta y Visualización**

A continuación, se presenta un script que conecta a la base de datos, ejecuta una Agregación (**Aggregation Framework**) para resumir la población por ciclo de vida y genera una visualización rápida. Esto simula lo que haría una herramienta de BI conectada a nuestro clúster.

```python
try:
    client = MongoClient(uri, tlsCAFile=certifi.where())
    collection = client["SisbenDB"]["Personas_Procesadas"]
    print("Conectado a MongoDB para análisis.")
except Exception as e:
    print(f"Error de conexión: {e}")

# Objetivo: Contar cuántas personas hay por 'Ciclo_Vital' ordenado alfabéticamente
pipeline = [
    {
        "$group": {
            "_id": "$Ciclo_Vital",
            "total_registros": {"$sum": 1}
        }
    },
    {
        "$sort": {"_id": 1}
    }
]

resultados = list(collection.aggregate(pipeline))
df_viz = pd.DataFrame(resultados)

df_viz.rename(columns={"_id": "Ciclo de Vida", "total_registros": "Cantidad"}, inplace=True)

print("Datos recuperados de Atlas:")
display(df_viz)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_viz, x="Cantidad", y="Ciclo de Vida", hue="Ciclo de Vida", palette="viridis", legend=False)

plt.title("Distribución de Población Sisbén por Ciclo Vital (Fuente: MongoDB Atlas)")
plt.xlabel("Cantidad de Registros")
plt.ylabel("Etapa del Ciclo Vital")
plt.grid(axis='x', linestyle='--', alpha=0.7)

for index, value in enumerate(df_viz["Cantidad"]):
    plt.text(value, index, f' {value:,}', va='center', fontsize=9)

plt.show()
```

### **4.1 Interpretación del Ejemplo**

El gráfico anterior demuestra la **eficacia del preprocesamiento**:

* **Orden Cronológico:** Gracias a que añadimos prefijos numéricos en Spark (`01.`, `02.`), el gráfico se ordena lógicamente por edad y no alfabéticamente.
* **Velocidad:** La consulta tardó **milisegundos** en traer el resumen, ya que no tuvo que leer los 2 millones de filas en local, sino que MongoDB procesó la suma en sus servidores.
* **Integridad:** Los datos visualizados provienen directamente de la fuente persistente, asegurando que cualquier usuario consulte la misma **"verdad única"**.

## **5. Limitaciones y Desafíos de Infraestructura**

Durante la fase de persistencia hacia la nube (MongoDB Atlas), se identificaron restricciones operativas relacionadas con la infraestructura **gratuita** utilizada para este prototipo.

---
### **5.1. Restricciones de la Capa Gratuita (MongoDB Atlas M0)**

Aunque el clúster M0 tiene capacidad de almacenamiento, impone límites estrictos de rendimiento para proteger los recursos compartidos:

* **Rate Limiting (Limitación de Tasa):** El clúster gratuito aplica un límite no documentado de operaciones de escritura por segundo (IOPS). Al inyectar 4.46 millones de documentos de forma paralela con Spark, Atlas comenzó a rechazar conexiones tras alcanzar un umbral de aproximadamente **1.9 millones de documentos**.
* **Throttling de Red:** Se observaron **tiempos de espera (timeouts)** crecientes en los *workers* de Spark, lo que indica que el clúster de destino forzó una desaceleración de la ingesta para mantener su estabilidad.


---
### **5.2. Impacto y Mitigación**

* **Datos Disponibles:** Se logró persistir exitosamente aproximadamente el **42% del dataset total (1.9 millones de 4.46 millones)**.
* **Validez del Piloto:** A pesar de no contar con la totalidad de los registros, la muestra de 1.9 millones es **estadísticamente significativa** y suficiente para validar completamente la arquitectura de la solución, los *pipelines* de agregación y las visualizaciones propuestas en este piloto.
* **Escalabilidad Futura:** En un entorno de producción, esta limitación se resolvería fácilmente migrando a un **clúster dedicado (M10 o superior)**. Esto eliminaría el *throttling* y permitiría la carga del 100% de los datos rápidamente usando el mismo código desarrollado.
